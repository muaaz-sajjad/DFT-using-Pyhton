```python
import numpy as np
import matplotlib.pyplot as plt

# =============================================
# 1. Improved Numerov Solver (UNCHANGED)
# =============================================
def numerov(k, u0, u1, dr):
    u = np.zeros_like(k)
    u[0] = u0
    u[1] = u1
    dr_sqr = dr**2
    for i in range(2, len(k)):
        c0 = 1 + (dr_sqr / 12) * k[i-2]
        c1 = 2 * (1 - (5 * dr_sqr / 12) * k[i-1])
        c2 = 1 + (dr_sqr / 12) * k[i]
        u[i] = (c1 * u[i-1] - c0 * u[i-2]) / c2
    return u

def solve_schrodinger(r, V, eps_min=-5, eps_max=0, maxiter=200, stoptol=1e-6):
    dr = r[1] - r[0]
    for _ in range(maxiter):
        eps = (eps_min + eps_max) / 2
        k = 2 * (eps - V)

        u0 = r[-1] * np.exp(-np.sqrt(-2*eps)*r[-1]) if eps < 0 else r[-1]*np.exp(-r[-1])
        u1 = r[-2] * np.exp(-np.sqrt(-2*eps)*r[-2]) if eps < 0 else r[-2]*np.exp(-r[-2])

        u = numerov(k[::-1], u0, u1, dr)[::-1]
        u /= np.linalg.norm(u)

        if abs(u[0]) < stoptol:
            return eps, u
        if (u[0] > 0 and eps_min == -5) or (u[0] < 0 and eps_max == 0):
            eps = (eps_min + eps_max)/2
        if u[0] > 0:
            eps_min = eps
        else:
            eps_max = eps
    raise Exception(f'Not converged after {maxiter} iterations (E∈[{eps_min:.3f},{eps_max:.3f}])')

# =============================================
# 2. Poisson Solver (UNCHANGED)
# =============================================
def verlet(f, U0, U1, dr):
    dr_sqr = dr**2
    U_H = np.zeros_like(f)
    U_H[0] = U0
    U_H[1] = U1
    for i in range(2, len(f)):
        U_H[i] = 2 * U_H[i-1] - U_H[i-2] + f[i-1] * dr_sqr
    return U_H

def solve_poisson(r, u):
    dr = r[1] - r[0]
    n = u**2 / (4 * np.pi * r**2)
    f = -4 * np.pi * r * n
    U_H = verlet(f, 0.0, dr * f[0], dr)
    return U_H / r

# =============================================
# 3. DFT-LDA (UNCHANGED except return values)
# =============================================
C2 = -0.738
Aa, B, C, D = 0.0311, -0.048, 0.002, -0.0116
gamma, beta1, beta2 = -0.1423, 1.0529, 0.3334

def exc(n):
    V_x = (4/3) * C2 * n**(1/3)
    e_x = C2 * n**(1/3)
    return V_x, e_x

def cor(n):
    rs = (3/(4*np.pi*n))**(1/3)
    V_c = np.piecewise(rs, [rs < 1, rs >= 1], [
        lambda rs: Aa*np.log(rs) + B - Aa/3 + (2/3)*C*rs*np.log(rs) + (2*D-C)*rs/3,
        lambda rs: gamma/(1 + beta1*np.sqrt(rs) + beta2*rs) *
                  (1 + (7/6)*beta1*np.sqrt(rs) + beta2*rs) /
                  (1 + beta1*np.sqrt(rs) + beta2*rs)
    ])
    e_c = np.piecewise(rs, [rs < 1, rs >= 1], [
        lambda rs: Aa*np.log(rs) + B + C*rs*np.log(rs) + D*rs,
        lambda rs: gamma/(1 + beta1*np.sqrt(rs) + beta2*rs)
    ])
    return V_c, e_c

def dft(V_en, maxiter=50, stop=1e-5, mixing=0.3, correlation=False, verbose=True):
    V_H = np.zeros_like(V_en)
    V_xc = np.zeros_like(V_en)
    e_xc = np.zeros_like(V_en)
    eps_prev = None
    n_prev = np.zeros_like(V_en)

    for i in range(maxiter):
        eps, u = solve_schrodinger(r, V_en + V_H + V_xc)
        u /= np.linalg.norm(u) * np.sqrt(r[1]-r[0])

        n = 2 * (u**2 / (4 * np.pi * r**2))
        n = mixing * n + (1-mixing) * n_prev

        if eps_prev is not None and abs(eps - eps_prev) < stop:
            E_total = 2*eps - np.trapz(V_H * n, r) - np.trapz((V_xc - e_xc) * n, r)
            if verbose: print(f"Converged after {i} iterations")
            return E_total, V_xc, n, u, eps, V_H

        U_H = solve_poisson(r, u)
        V_H = 2 * U_H / r

        V_x, e_x = exc(n)
        if correlation:
            V_c, e_c = cor(n)
            V_xc, e_xc = V_x + V_c, e_x + e_c
        else:
            V_xc, e_xc = V_x, e_x

        eps_prev = eps
        n_prev = n

        if verbose:
            print(f"Iter {i:2d}: E = {eps:.6f}, ΔE = {abs(eps-eps_prev):.2e}"
                  if eps_prev else f"Iter {i:2d}: E = {eps:.6f}")

    raise Exception(f"No convergence after {maxiter} iterations")

# =============================================
# 4. NEW: Radii Calculation Function (ADDED)
# =============================================
def calculate_radii(r, n):
    """Compute <r>, r_mp, and r_rms from electron density"""
    dr = r[1] - r[0]
    r_avg = np.trapz(r * n, r)          # <r> expectation value
    r_mp = r[np.argmax(n)]              # Most probable radius
    r_rms = np.sqrt(np.trapz(r**2 * n, r))  # RMS radius
    return {"<r>": r_avg, "r_mp": r_mp, "r_rms": r_rms}

# =============================================
# 5. Modified Plotting Function (UPDATED)
# =============================================
def plot_results(r, V_en, V_H, V_xc, n, u, eps, Z=2):
    """Now includes radius markers"""
    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))

    # 1. Potentials (UNCHANGED)
    ax1.plot(r, V_en, 'r-', label='Nuclear $V_{en}$')
    ax1.plot(r, V_H, 'b--', label='Hartree $V_H$')
    ax1.plot(r, V_xc, 'g:', label='XC $V_{xc}$')
    ax1.plot(r, V_en + V_H + V_xc, 'k-', lw=2, label='Total Potential')
    ax1.set_title('Potentials')
    ax1.set_ylabel('Energy (Ha)')
    ax1.legend()
    ax1.grid(True)

    # 2. Density/Wavefunction (UPDATED WITH RADII MARKERS)
    ax2.plot(r, n, 'm-', label='Electron Density $n(r)$')
    ax2.plot(r, u, 'c--', label='Wavefunction $u(r)$')

    # NEW: Add radius markers
    radii = calculate_radii(r, n)
    ax2.axvline(radii['<r>'], color='k', linestyle=':', label=f'<r> = {radii["<r>"]:.2f} bohr')
    ax2.axvline(radii['r_mp'], color='b', linestyle='--', label=f'r_mp = {radii["r_mp"]:.2f} bohr')

    ax2.set_title('Electron Distribution')
    ax2.set_ylabel('$n(r)$ / $u(r)$')
    ax2.legend()
    ax2.grid(True)

    # 3. Error Analysis (UNCHANGED)
    n_exact = (2 * Z**3 / np.pi) * np.exp(-2 * Z * r)
    ax3.semilogy(r, abs(n - n_exact), 'k-', label='Density Error')
    ax3.set_title('Convergence Check')
    ax3.set_xlabel('Radius (Bohr)')
    ax3.set_ylabel('Absolute Error')
    ax3.legend()
    ax3.grid(True)

    plt.tight_layout()
    plt.show()

# =============================================
# 6. Main Execution (UPDATED WITH RADII OUTPUT)
# =============================================
r = np.linspace(1e-6, 15, 5001)[1:]  # Finer grid near nucleus
Z = 2
V_en = -Z / r

print("=== Helium Atom DFT Calculation ===")

# Without correlation (UPDATED to capture radii)
print("\n1. Without correlation:")
E_no_corr, V_xc_no, n_no, u_no, eps_no, V_H_no = dft(V_en, correlation=False)
radii_no = calculate_radii(r, n_no)
plot_results(r, V_en, V_H_no, V_xc_no, n_no, u_no, eps_no)

# With correlation (UPDATED to capture radii)
print("\n2. With correlation:")
E_with_corr, V_xc_with, n_with, u_with, eps_with, V_H_with = dft(V_en, correlation=True)
radii_with = calculate_radii(r, n_with)
plot_results(r, V_en, V_H_with, V_xc_with, n_with, u_with, eps_with)

# NEW: Comprehensive results output
print("\n=== Final Results ===")
print(f"{'Quantity':<15} {'No Correlation':<20} {'With Correlation':<20}")
print(f"{'-'*55}")
print(f"{'Energy (Ha)':<15} {E_no_corr:<20.6f} {E_with_corr:.6f}")
print(f"{'<r> (bohr)':<15} {radii_no['<r>']:<20.6f} {radii_with['<r>']:.6f}")
print(f"{'r_mp (bohr)':<15} {radii_no['r_mp']:<20.6f} {radii_with['r_mp']:.6f}")
print(f"{'r_rms (bohr)':<15} {radii_no['r_rms']:<20.6f} {radii_with['r_rms']:.6f}")
print(f"\nCorrelation energy: {E_with_corr-E_no_corr:.6f} Ha")
```
