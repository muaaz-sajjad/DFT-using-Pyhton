```python
# =============================
# STEP 01: IMPORTING LIBRARIES
#==============================
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ===================================================================================
# STEP 02: NUMEROV SOLVE (Solving Radial Part of Schrodinger equation for 1s orbital)
#====================================================================================
x_min = 0
x_max = np.pi
h = 0.1 # step size
x = np.arange(x_min, x_max + h, h)
N = len(x)

# step 1: define first step
# f(x) = -1 for all x
f = -1 * np.ones(N)
print(x)


# step 2: initialize y-array

y = np.zeros(N)
y[0] = 0
y[1] = np.sin(h)
print(f'y(0): {y[0]}')
print(f'y(1): {y[1]}')

for i in range(1, N-1):
  y[i+1] =( (2 * (1 - (5 * h**2 * f[i]) / 12) * y [i]) - (1 + (h**2 * f[i-1]) / 12) * y [i-1] ) / (1 + ((h**2 * f[i+1]) / 12))

y_exact = np.sin(x)
plt.figure(figsize=(8,5))
plt.plot(x, y, label='Numerov Solution', linewidth=2)
plt.plot(x, y_exact, '--', label='Analytical Solution (sin(x))', linewidth=2)
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Numerov Method vs Analytical Solution for y'' = -y")
plt.legend()
plt.grid(True)
plt.show()
```
